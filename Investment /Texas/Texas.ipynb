```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
population = pd.read_csv("Resources/TXPOP.csv")

population = population.rename(columns={'TXPOP': 'population'})

# Review the DataFrame
population
```

```python
# Convert 'date' column to datetime format
population['DATE'] = pd.to_datetime(population['DATE'])

# Grouping by year extracted from the 'date' column and calculating the sum of 'Value' for each year
population['DATE'] = population['DATE'].dt.year  # Extracting year
population = population.rename(columns={'DATE': 'year'})

# Delete the years to match the other data
population = population.drop(population[population['year'] < 2016].index)

population
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
employment = pd.read_csv("Resources/TXNA.csv")

employment = employment.rename(columns={'TXNA': 'employment'})

# Review the DataFrame
employment
```

```python
# Convert 'date' column to datetime format
employment['DATE'] = pd.to_datetime(employment['DATE'])

# Grouping by year extracted from the 'date' column and calculating the sum of 'Value' for each year
employment['year'] = employment['DATE'].dt.year  # Extracting year
employment = employment.groupby('year')['employment'].mean().reset_index()

# Delete the years to match the other data
employment = employment.drop(employment[employment['year'] < 2016].index)

employment
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
income = pd.read_csv("Resources/INCOME.csv")

income = income.rename(columns={'MEHOINUSTXA646N': 'income'})

# Review the DataFrame
income
```

```python
# Convert 'date' column to datetime format
income['DATE'] = pd.to_datetime(income['DATE'])

# Grouping by year extracted from the 'date' column and calculating the sum of 'Value' for each year
income['DATE'] = income['DATE'].dt.year  # Extracting year
income = income.rename(columns={'DATE': 'year'})

# Delete the years to match the other data
income = income.drop(income[income['year'] < 2016].index)

# Display the dataframe
income

#(thousand of dollars/year)
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
home_size = pd.read_csv("Resources/MEDSQUFEETX.csv")

home_size = home_size.rename(columns={'MEDSQUFEETX': 'home_size(sqft)'})

# Review the DataFrame
home_size
```

```python
# Convert 'date' column to datetime format
home_size['DATE'] = pd.to_datetime(home_size['DATE'])

# Grouping by year extracted from the 'date' column and calculating the sum of 'Value' for each year
home_size['year'] = home_size['DATE'].dt.year  # Extracting year
home_size = home_size.groupby('year')['home_size(sqft)'].mean().reset_index()

# Delete the year 2023 to match the other data
home_size = home_size.drop(home_size[home_size['year'] == 2023].index)

home_size
```

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
house_prices = pd.read_csv("Resources/TXSTHPI.csv")

house_prices = house_prices.rename(columns={'TXSTHPI': 'house_prices'})

# Review the DataFrame
house_prices
```

```python
# Convert 'date' column to datetime format
house_prices['DATE'] = pd.to_datetime(house_prices['DATE'])

# Grouping by year extracted from the 'date' column and calculating the sum of 'Value' for each year
house_prices['year'] = house_prices['DATE'].dt.year  # Extracting year
house_prices = house_prices.groupby('year')['house_prices'].mean().reset_index()

house_prices
```

```python
# Merging population and employment DataFrames on 'year' column
merged_df = pd.merge(population, employment, on='year')

# Merging income DataFrame to the already merged DataFrame
merged_df = pd.merge(merged_df, income, on='year')

# Merging home_size DataFrame to the already merged DataFrame
merged_df = pd.merge(merged_df, home_size, on='year')

# Merging house_prices DataFrame to the already merged DataFrame
merged_df = pd.merge(merged_df, house_prices, on='year')

merged_df
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting the first set of values on the primary y-axis (left)
ax1.plot(merged_df['year'], merged_df['population'], marker='o', markersize=4, color='lightcoral', label='population')
ax1.set_xlabel('Year')
ax1.tick_params(axis='y', labelcolor='lightcoral')

# Creating secondary y-axes (right) for other sets of values
ax2 = ax1.twinx()
ax2.plot(merged_df['year'], merged_df['employment'], marker='o', markersize=4, color='orange', label='employment')
ax2.tick_params(axis='y', labelcolor='red')

ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 80))  # Adjust the position of the new axis
ax3.plot(merged_df['year'], merged_df['income'], marker='o', markersize=4, color='yellowgreen', label='income')
ax3.tick_params(axis='y', labelcolor='green')

ax4 = ax1.twinx()
ax4.spines['right'].set_position(('outward', 160))  # Adjust the position of the new axis
ax4.plot(merged_df['year'], merged_df['home_size(sqft)'], marker='o', markersize=4, color='turquoise', label='home size(sqft)')
ax4.tick_params(axis='y', labelcolor='purple')

ax5 = ax1.twinx()
ax5.spines['right'].set_position(('outward', 240))  # Adjust the position of the new axis
ax5.plot(merged_df['year'], merged_df['house_prices'], marker='o', markersize=4, color='slateblue', label='house prices')
ax5.tick_params(axis='y', labelcolor='orange')

# Adding legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
lines4, labels4 = ax4.get_legend_handles_labels()
lines5, labels5 = ax5.get_legend_handles_labels()

plt.title('Texas')
plt.show()
```
